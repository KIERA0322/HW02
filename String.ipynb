```python
#Make a random string that has enough variation to experiment on.
Honeycrisp = '12\nedison\n\tgobeldygook\nDOES ANYONE even look at these?'
print(Honeycrisp)
```

```python
#Capitalize and lower the string
print(Honeycrisp.lower())
print(Honeycrisp.upper())
```

```python
#Print the length of the string
print(len(Honeycrisp))
```

```python
#Split the string at the k
print(Honeycrisp.split("k"))
```

```python
#Split the string at any whitespace
print(Honeycrisp.split())
```

```python
#Replace the k's with some bs
print(Honeycrisp.replace('k','ooooo'))
```
